## GEOPY - Mencari Letak Koordinat Wilayah dengan Geopy

```python
#memasukan modul yang digunakan
import geopy
import pandas as pd, os
```

```python
#memasukan data yang digunakan
df = pd.read_csv("D:/Data Penelitian/2.csv" , sep=',')
df
```

```python
#memanggil fungsi nominatim
from geopy.geocoders import Nominatim
nom = Nominatim()
```

```python
#untuk melihat letak geografis di google map (latitude,longitude)
n=nom.geocode("London eye")
print(n.latitude, n.longitude)
n=nom.geocode("Jakarta")
print(n.latitude, n.longitude)
n=nom.geocode("Eifel tower")
print(n.latitude, n.longitude)
n=nom.geocode("indonesia")
print(n.latitude, n.longitude)
n=nom.geocode("malaysia")
print(n.latitude, n.longitude)
n=nom.geocode("jawa tengah")
print(n.latitude, n.longitude)
```

```python
#untuk melihat latitude dan longitude kota karawang
n=nom.geocode("Karawang")
print(n.latitude, n.longitude)
```

```python
#add a new address column (tambahan kolom alamat yang baru). kolom alamat baru tersebut diperoleh dari penggabungan kolom alamat, city, name dan country
df["Address"]= df["Address"]+", "+df["City"]+", "+df["Name"]+", "+df["Country"]
df
```

```python
#menambahkan kolom koordinat dan menampilkan letak koordinat
#Send this string to Geocode method
df["Coordinate"]=df["Name"].apply(nom.geocode)
df
```

```python
#menambahkan kolom latitude dan menampilkan letak/nilai latitude
df["Latitude"]=df["Coordinate"].apply(lambda x: x.latitude if x !=None else None)
df
```

```python
#menambahkan kolom longitude dan menampilkan letak/nilai longitude 
df["Longitude"]=df["Coordinate"].apply(lambda x: x.longitude if x !=None else None)
df
```
